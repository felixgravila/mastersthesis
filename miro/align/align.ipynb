```python
import random

def generate_dna(l):
    dna = []
    for i in range(l):
        r = round(random.uniform(0,3))
        dna.append(r)
    return dna
```

```python
class Recurse(Exception):
    def __init__(self, *args, **kwargs):
        self.args = args
        self.kwargs = kwargs

def recurse(*args, **kwargs):
    raise Recurse(*args, **kwargs)
        
def tail_recursive(f):
    def decorated(*args, **kwargs):
        while True:
            try:
                return f(*args, **kwargs)
            except Recurse as r:
                args = r.args
                kwargs = r.kwargs
                continue
    return decorated
```

```python
#use recursive instead
@tail_recursive
def calc_score_rec(dna1, dna2, s1_from, s2_from, overlap_length, total_score, with_print = False):

    if(overlap_length == 0):
        return total_score, dna1, dna2
      
    score = -1
    if(dna1[s1_from] == dna2[s2_from]):
        score = 1
    
    if(with_print):
        print(f"Rec; DNA1:{dna1}, DNA1:{dna2}, Index1:{s1_from},Index2:{s2_from}, val1 = {dna1[s1_from]}, val2 = {dna2[s2_from]} ,overlap={overlap_length}, total:{total_score}")

    if(score == -1 and len(dna1) > s1_from + 2 and len(dna2) > s2_from +1):
        if(dna1[s1_from+1] == dna2[s2_from] and dna1[s1_from+2] == dna2[s2_from+1]):
            new_dna2 = dna2[:s2_from] + [4] + dna2[s2_from:]
            overlap_length -= 1
            return calc_score_rec(dna1, new_dna2, s1_from + 1, s2_from + 1, overlap_length, total_score, with_print)
    
    if(score == -1 and len(dna2) > s2_from + 2 and len(dna1) > s1_from +1):
        if(dna1[s1_from] == dna2[s2_from+1] and dna1[s1_from+1] == dna2[s2_from+2]):
            new_dna1 = dna1[:s1_from] + [4] + dna1[s1_from:]
            overlap_length -= 1
            return calc_score_rec(new_dna1, dna2, s1_from + 1, s2_from + 1, overlap_length, total_score, with_print)
                    
    return calc_score_rec(dna1, dna2, s1_from + 1, s2_from + 1, overlap_length - 1, total_score + score, with_print) 
```

```python
def calc_score_it(dna1, dna2, s1_from, s2_from, overlap_length, total_score, with_print = False):
    
    while(overlap_length > 0):
        
        score = -1
        if(dna1[s1_from] == dna2[s2_from]):
            score = 1
            
        if(with_print):
            print(f"It; DNA1:{dna1}, DNA1:{dna2}, Index1:{s1_from},Index2:{s2_from}, val1 = {dna1[s1_from]}, val2 = {dna2[s2_from]} ,overlap={overlap_length}, total:{total_score}")

        if(score == -1 and len(dna1) > s1_from + 2 and len(dna2) > s2_from +1):
            if(dna1[s1_from+1] == dna2[s2_from] and dna1[s1_from+2] == dna2[s2_from+1]):
                dna2 = dna2[:s2_from] + [4] + dna2[s2_from:]
                s1_from += 1
                s2_from += 1
                overlap_length -= 1
                #total_score += 0.5
                continue
        
        if(score == -1 and len(dna2) > s2_from + 2 and len(dna1) > s1_from +1):
            if(dna1[s1_from] == dna2[s2_from+1] and dna1[s1_from+1] == dna2[s2_from+2]):
                dna1 = dna1[:s1_from] + [4] + dna1[s1_from:]
                s1_from += 1
                s2_from += 1
                overlap_length -= 1
                #total_score += 0.5
                continue
        
        s1_from += 1
        s2_from += 1
        total_score += score
        overlap_length -= 1

    return total_score, dna1, dna2
```

```python
def get_best_alignment_rec(dna1, dna2, with_print = False):
    max_score = None
    max_score_s1_i = -1 
    max_score_s2_i = -1
    max_dna1 = []
    max_dna2 = []

    overlap = 1
    s1_i = 0
    s2_i = len(s2) - 1
    max_overlap = min(len(s1), len(s2))

    while(overlap > 0): 
        
        #score, new_dna1, new_dna2 = calc_score_rec(s1,s2,s1_i, s2_i, overlap, 0, with_print)
        score, new_dna1, new_dna2 = calc_score_it(s1,s2,s1_i, s2_i, overlap, 0, with_print)

        if(with_print):
            print(f"overlap: {overlap}")
            print(f"score: {score}")

        if(max_score is None or score > max_score):
            max_score = score 
            max_score_s1_i = s1_i
            max_score_s2_i = s2_i
            max_dna1 = new_dna1
            max_dna2 = new_dna2

        if(s1_i < s2_i):
            s2_i -= 1
            overlap = min(len(s1), len(s2) - s2_i)
        else: 
            s1_i += 1
            overlap = min(len(s1) - s1_i, len(s2))
    
    if(with_print):
        print(f"max score: {max_score}")
    return max_score, max_score_s1_i, max_score_s2_i, max_dna1, max_dna2
```

```python
def visualize_alignment(s1, s2, s1_i, s2_i):
    s1_prefix = 0
    s2_prefix = 0
    s1_suffix = 0
    s2_suffix = 0
        
    if(s2_i > s1_i):
        s1_prefix = s2_i
    elif(s2_i < s1_i):
        s2_prefix = s1_i
        
    bases = ['A', 'T', 'G', 'C', '_']
    
    string_1 = s1_prefix * "_"
    string_2 = s2_prefix * "_"
    
    for i in range(len(s1)):
        val = s1[i]
        string_1 += bases[val]
            
    for i in range(len(s2)):
        val = s2[i]
        string_2 += bases[val]
        
    if(len(string_1) > len(string_2)):
        string_2 += (len(string_1) - len(string_2)) * "_"
    elif(len(string_2) > len(string_1)):
        string_1 += (len(string_2) - len(string_1)) * "_"
        
    print(string_1)
    #print("-------")
    print(string_2)
    
def to_base_string(s):
    base_s = ""
    bases = ['A', 'T', 'G', 'C', '_']
    for i in range(len(s)):
        val = s[i]
        base_s += bases[val]
    return base_s

def to_base_list(base_s):
    s = []
    for i in range(len(base_s)):
        b = base_s[i]
        if(b == 'A'):
            s.append(0)
        elif(b == 'T'):
            s.append(1)
        elif(b =='G'):
            s.append(2)
        elif(b =='C'):
            s.append(3)
        elif(b=='_'):
            s.append(4)
        else:
            raise Exception(f"Invalid base string, contains character {b}.")
    return s
```

```python
def pad_sequences(s1, s2, s1_i, s2_i):
    s1_prefix = 0
    s2_prefix = 0
    s1_suffix = 0
    s2_suffix = 0
    
    if(s2_i > s1_i):
        s1_prefix = s2_i
    elif(s2_i < s1_i):
        s2_prefix = s1_i
    
    s1_padded = []
    s2_padded = []
    for i in range(s1_prefix):
        s1_padded.append(4)
    for i in range(s2_prefix):
        s2_padded.append(4)
    
    s1_padded.extend(s1)
    s2_padded.extend(s2)
                
    if(len(s1_padded) > len(s2_padded)):
        for i in range(len(s1_padded) - len(s2_padded)):
            s2_padded.append(4)
    elif(len(s2_padded) > len(s1_padded)):
        for i in range(len(s2_padded) - len(s1_padded)):
            s1_padded.append(4)  
    return s1_padded, s2_padded   
```

```python
def print_s(s1,s2):
    print(s1)
    print(s2)
    print("---")

s1 = to_base_list("GGGGTAGTATCCCACCTTTTCCCGAGGCAACATC")
s2 = to_base_list("TAGTATCCCACGCTTTTTCCAGGCAAACGTCGC")
score, i1, i2, dna1, dna2 = get_best_alignment_rec(s1,s2,False)
s1, s2 = pad_sequences(dna1,dna2,i1,i2)
print_s(s1,s2)

s1 = to_base_list("TAGTATCCCACCTTTTCCCGAGGCAACATC")
s2 = to_base_list("TAGTATCCCACGCTTTTTCCAGGCAAACGTCGC")
score, i1, i2, dna1, dna2 = get_best_alignment_rec(s1,s2,False)
s1, s2 = pad_sequences(dna1,dna2,i1,i2)
print_s(s1,s2)

s1 = to_base_list("GGGGTAGTATCCCACCTTTTCCCGAGGCAACATC")
s2 = to_base_list("TAGTATCCCACGCTTTTTCCAGGCAAACGTCGCCCC")
score, i1, i2, dna1, dna2 = get_best_alignment_rec(s1,s2,False)
s1, s2 = pad_sequences(dna1,dna2,i1,i2)
print_s(s1,s2)
```

```python
def pad_ends(aligned):
    m = 0
    for i in range(len(aligned)):
        if(len(aligned[i])>m):
            m = len(aligned[i])    
    aligned_padded = []
    for i in range(len(aligned)):
        len_to_pad = m - len(aligned[i])
        element = aligned[i] + (len_to_pad * [4])
        aligned_padded.append(element)
        
    return aligned_padded

        
aligned_padded = pad_ends(aligned)
```

```python
import collections

np_ap = np.array(aligned_padded)
np_ap = np_ap.transpose()

rows = np_ap.shape[0]
cols = np_ap.shape[1]

dna = []
for i in range(rows):
    common = collections.Counter(np_ap[i]).most_common(2)
    if(common[0][0] == 4 and len(common) == 1): #this should not technically happen
        dna.append(4)
    elif(common[0][0] == 4 and len(common) > 1):
        dna.append(common[1][0])
    else:
        dna.append(common[0][0])

#print("Assembled DNA:")
#print(to_base_string(dna))
        
```

```python
def lcs(X, Y): 
    # find the length of the strings 
    m = len(X) 
    n = len(Y) 
  
    # declaring the array for storing the dp values 
    L = [[None]*(n + 1) for i in range(m + 1)] 
  
    """Following steps build L[m + 1][n + 1] in bottom up fashion 
    Note: L[i][j] contains length of LCS of X[0..i-1] 
    and Y[0..j-1]"""
    for i in range(m + 1): 
        for j in range(n + 1): 
            if i == 0 or j == 0 : 
                L[i][j] = 0
            elif X[i-1] == Y[j-1]: 
                L[i][j] = L[i-1][j-1]+1
            else: 
                L[i][j] = max(L[i-1][j], L[i][j-1]) 
  
    # L[m][n] contains the length of LCS of X[0..n-1] & Y[0..m-1] 
    return L[m][n] 
# end of function lcs 

#Test1
X = "AGGTABBB"
Y = "GXTXAYBBB"
print("LCS is ", lcs(X, Y))
print("---")

#Test2
s1 = to_base_list("GTAGTATCCCACCTTTTCCCGAGGCAACATC")
s2 = to_base_list("TAGTATCCCACGCTTTTTCCAGGCAAACGTCGC")
score, i1, i2, dna1, dna2 = get_best_alignment_rec(s1,s2,False)
print_test(dna1,dna2,i1,i2)
lcs_res = lcs(s1,s2)

print(f"DNA length:{len(s1)}")
print(f"Reference length:{len(s2)}")
print(f"LCS is:{lcs_res}")
print("---")
print("---")
print("---")

#Run
s1 = dna
s2 = to_base_list(reference)
lcs_res = lcs(s1,s2)

print(f"DNA length:{len(s1)}")
print(f"Reference length:{len(s2)}")
print(f"LCS is:{lcs_res}")
```
